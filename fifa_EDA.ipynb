```python
import numpy as np
import pandas as pd
```

```python
fifa_data = pd.read_csv('/content/fifa_eda.csv')
fifa_data.head()
```

```python
fifa_data.info()
```

```python
fifa_data.dropna(inplace = True)
```

```python
fifa_data.info()
```

```python
fifa_data.describe()
```

```python
fifa_data.describe(include = 'object')
```

```python
# Most frequent nationality
most_frequent_nationality = fifa_data['Nationality'].mode()
most_frequent_nationality
```

```python
# Distribution of player positions
position_distribution = fifa_data['Position'].value_counts()
position_distribution
```

```python
# Club with the highest average wage
average_wage_by_club = fifa_data.groupby('Club')['Wage'].mean().sort_values(ascending = False)
average_wage_by_club.head()
```

```python
#Which position has the highest average potential
average_potential_by_position = fifa_data.groupby('Position')['Potential'].mean().sort_values(ascending=False)
average_potential_by_position.head()
```

```python
#Which club has the highest total player value
total_value_by_club = fifa_data.groupby('Club')['Value'].sum().sort_values(ascending=False)
total_value_by_club
```

```python
#Which nationality has the highest average age
average_age_by_nationality = fifa_data.groupby('Nationality')['Age'].mean().sort_values(ascending=False)
average_age_by_nationality
```

```python
#Which club has the highest average international reputation
average_reputation_by_club = fifa_data.groupby('Club')['International Reputation'].mean().sort_values(ascending=False)
average_reputation_by_club.head(10)
```

```python
#Which 5 clubs have the most players with a Potential above 85
top_Potential_clubs = fifa_data[fifa_data['Potential'] > 85]['Club'].value_counts().head(5)
top_Potential_clubs
```

```python
#Which clubs have the most diverse representation of nationalities among their players
club_nationality_diversity = fifa_data.groupby('Club')['Nationality'].nunique().sort_values(ascending=False)
club_nationality_diversity
```

```python
# How are skill moves distributed among players in different positions (e.g., 'ST', 'CAM')
skill_moves_by_position = fifa_data.groupby('Position')['Skill Moves'].value_counts()
skill_moves_by_position
```

```python
fifa_data.to_csv('fifa_clean.csv', index = False)
```
